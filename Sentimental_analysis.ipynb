```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

# MODEL

```python
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')

model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

# CALCULATE SENTIMENT

```python
tokens=tokenizer.encode('Well, I kinda hate to admit it but I like going to Apple Bees. I enjoy the atmosphere (louder than average) sports theme with reasonable prices. Drink and appetizers specials during football season (I am a huge fan). But this location is off the Hook! Views, service, food is great, love it!',return_tensors='pt')
```

```python
tokens
```

```python
result=model(tokens)
```

```python
result
```

```python
result.logits
```
